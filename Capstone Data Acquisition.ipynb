### This function receives a tv show title and finds the imdb id associated with it

```python
def check_and_get_imdb_id(title):
#     print "sleeping...."
    print title
    sleep(2)
    buul = False
    check = 0
    while (buul == False):
        if check >= 10:
#             print "unknown"
            return "unknown"
            buul = True
        elif imdb.get_title_by_id(imdb.search_for_title(title)[check]['imdb_id']).type == 'tv_series':
#             print "it's done"
#             print imdb.search_for_title(title)[check]['imdb_id']
            return imdb.search_for_title(title)[check]['imdb_id']
            buul = True
        else:
            check +=1
#             print "trying again"
```
