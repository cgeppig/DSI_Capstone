## Scraping shows from wikipedia

```python
urlwiki = urllib2.urlopen('https://en.wikipedia.org/wiki/List_of_American_television_programs_by_date')
```

```python
wikisoup = BeautifulSoup(urlwiki)
```

```python
wikishows = []
for div in wikisoup('div', {'id': 'mw-content-text'}):
    ## need to get out of 'div' id='toc'
    for ul in div('ul'):
        for li in ul('li'):
            for a in li('a'):
                wikishows.append(a.get_text())
#                 print a.get_text()
```
